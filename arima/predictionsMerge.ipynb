```python
import pandas as pd
import numpy as np
```

```python
for server in range(0,3):
    
    

    
    costFile = './predictions/predictions_server_' + str(server+1) + '_tactic_1_cost.csv'

    latencyFile = './predictions/predictions_server_' + str(server+1) + '_tactic_1_latency.csv'
    testFileName = "../parse_tactics_weekend_sample_rate/normalized_tva_server_"+ str(server+1) + "_tactic_1_test_weekdayOneHot.csv"
    
    data = pd.read_csv(testFileName)

    latencyData = pd.read_csv(latencyFile)
    costData = pd.read_csv(costFile)

    data['predicted_latency'] = latencyData['predicted_latency']
    data['predicted_cost'] = costData['predicted_cost']
    data['predicted_reliability'] = data['reliability']

    finalFile = '../predictions/arima_predictions_server_' + str(server+1) + '_tactic_1.csv'

    data.to_csv(finalFile, index = False)
```
