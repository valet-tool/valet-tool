```python
import pandas as pd
import numpy as np
from sklearn import preprocessing

data = pd.read_csv("document1.csv",parse_dates= ['Timestamp'])

data['Timestamp'] = pd.to_datetime(data['Timestamp'])
```

```python
data.head()
```

```python
data_new = data
```

```python
data_new.head()
```

```python
doW = data_new['Timestamp'].dt.day_name()
```

```python
hh = data_new['Timestamp'].dt.hour
```

```python
mm = data_new['Timestamp'].dt.minute
```

```python
ss = (data_new['Timestamp'].dt.second.astype(str)+"."+data_new['Timestamp'].dt.microsecond.astype(str)).astype(float)
```

```python
dataTimeCols = pd.concat([doW,hh,mm,ss],axis = 1,keys=['doW','Hours','Minutes','Seconds'])
```

```python
data_new = pd.concat([pd.DataFrame([0]*data_new.shape[0]), dataTimeCols, data_new.iloc[:,2:-1], pd.DataFrame([0]*data_new.shape[0])],axis = 1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_new[['Hours','Minutes','Seconds','Server','Tactic','Latency','Cost']] = scaler.fit_transform(data_new[['Hours','Minutes','Seconds','Server','Tactic','Latency','Cost']])
```

```python
one_hot = pd.get_dummies(data_new['doW'])
```

```python
data_new = data_new.drop('doW',axis = 1)
data_new = data_new.join(one_hot)
```

```python
data_new.head()
```

```python
data_new=data_new.drop(data_new.columns[0], axis=1)
```

```python
data_new['newCol'] = data_new.apply(lambda x: 0, axis=1)
```

```python
from sklearn.model_selection import train_test_split
A = (data_new) 
b = (data_new)
A_train, A_test, b_train, b_test = train_test_split(A, b, test_size=0.2, random_state=100)

```

```python
A_val_train,A_val_test, b_val_train,b_val_test = train_test_split(A_train, b_train, test_size=0.2, random_state=100)

```
